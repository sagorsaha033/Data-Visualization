```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data for the Radar Chart
labels = ['A', 'B', 'C', 'D', 'E']
values = [4, 3, 2, 5, 4]

# Number of variables
num_vars = len(labels)

# Compute angle of each axis in radians
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the loop for a closed radar chart
values += values[:1]
angles += angles[:1]

# Initialize the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot data
ax.plot(angles, values, color='dodgerblue', linewidth=2, marker='o', label='Values')
ax.fill(angles, values, color='dodgerblue', alpha=0.3)

# Set category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)

# Set r-labels (radial ticks)
ax.set_yticks(range(1, max(values)))
ax.set_yticklabels(map(str, range(1, max(values))), fontsize=10)
ax.set_ylim(0, max(values))

# Add grid styling
ax.spines['polar'].set_visible(False)
ax.grid(color='gray', linestyle='dotted', linewidth=1)

# Add title
plt.title('Radar Chart Example', size=14, y=1.1)

# Show the plot
plt.tight_layout()
plt.show()
```
