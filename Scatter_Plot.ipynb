```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5, 6, 7, 8]  # X-axis values
y = [2, 3, 5, 7, 6, 8, 9, 10]  # Y-axis values

# Create the connected scatter plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, '-o', color='blue', label='Data')  # Line and markers
plt.scatter(x, y, color='red', label='Points')  # Highlighting points

# Add labels and title
plt.title('Connected Scatter Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
