```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D plotting toolkit
import numpy as np
```

```python
# Generate random data for the 3D scatter plot
np.random.seed(42)  # For reproducibility
x = np.random.rand(50) * 10  # Random x values between 0 and 10
y = np.random.rand(50) * 10  # Random y values between 0 and 10
z = np.random.rand(50) * 10  # Random z values between 0 and 10
colors = np.random.rand(50)  # Random colors for the points

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plotting the scatter
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis', s=50)

# Add color bar for reference
color_bar = plt.colorbar(scatter, ax=ax)
color_bar.set_label("Color Intensity")

# Set labels and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Scatter Plot")

# Show the plot
plt.show()
```
