```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)  # 1000 data points from a normal distribution

# Create a histogram and get the values (bin edges and heights)
count, bins, ignored = plt.hist(data, bins=30, alpha=0.5, edgecolor='black')

# Calculate the midpoints of the bins
bin_centers = (bins[:-1] + bins[1:]) / 2

# Plot the frequency polygon by connecting the bin centers
plt.plot(bin_centers, count, marker='o', color='b', linestyle='-', linewidth=2)

# Add title and labels
plt.title('Frequency Polygon')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()
```
